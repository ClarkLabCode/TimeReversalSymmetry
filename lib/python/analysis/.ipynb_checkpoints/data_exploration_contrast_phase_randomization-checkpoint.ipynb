```python
contrast_collect = []
contrast_collect_contrast = []
nat_numbers = np.load('/mnt/d/data/panoramic/natural_scenes_numbers.npy')

for ii in range(len(nat_numbers)):
    img = np.load(folder_nat + f'natural_{nat_numbers[ii]}.npy')
    K_row = img.shape[0] 
    K_col = img.shape[1] 

    shift = 0
    img_roll = np.roll(img, shift, axis=1)

    ####### phase randomized
    img_ifft = hpfn.get_phase_randomized(img, rand_type='Gaussian' , delete_zero_freq=True)
    contrast_collect.extend(np.real(img_ifft.flatten()))

    # Spacial filtering
    pix_per_deg = K_col / 360
    FWHM = 5 # in degree
    sigma_for_gaussian = np.round(FWHM/(2*np.sqrt(2*np.log(2)))*pix_per_deg, 1) # smoothing gaussian
    pad_size = int(4*sigma_for_gaussian) # this comes from the fact that the gaussian is truncated at 4*std
    img_filtered = hpfn.get_filtered_spacial(np.real(img_ifft), pad_size, sigma_for_gaussian)

    # resizing
    img_resized = hpfn.get_resized(img_filtered)

    # contrast
    scope = 'full'
#     img_contrast = hpfn.get_contrast(img_resized.reshape((1, )+img_resized.shape), scope)
    img_contrast = img_resized / abs(img_resized).max()
    
    contrast_collect_contrast.extend(img_contrast.flatten())
```

```python
fig = plt.figure(figsize=(12, 3), constrained_layout=True)
gs = GridSpec(1, 4, figure=fig)

hist, bin_centers = hpfn.get_hist(np.array(contrast_collect))

ax = fig.add_subplot(gs[0, 0])
ax.plot(bin_centers, hist, alpha=1.)
# plt.xlim([-100, 100])
ax.set_yscale('log')
ax.set_xlabel('luminance')
ax.set_ylabel('density')

hist, bin_centers = hpfn.get_hist(np.array(contrast_collect_contrast))
mean1 = np.round(np.array(contrast_collect_contrast).mean(), 2)
std1 = np.round(np.array(contrast_collect_contrast).std(), 2)
skew1 = np.round(skew(np.array(contrast_collect_contrast)), 2)
kurtosis1 = np.round(kurtosis(np.array(contrast_collect_contrast)), 2)

ax = fig.add_subplot(gs[0, 1])
ax.plot(bin_centers, hist, alpha=1.)
# plt.xlim([-100, 100])
ax.set_yscale('log')
ax.set_xlabel('contrast')
ax.set_ylabel('density')
ax.set_title(str(mean1) + ', ' + str(std1) + ', ' + str(skew1) + ', ' + str(kurtosis1), fontsize=10)

img_gaussian = np.random.normal(loc=0., scale=1, size=img.shape)
hist, bin_centers = hpfn.get_hist(np.real(img_gaussian))

ax = fig.add_subplot(gs[0, 2])
ax.plot(bin_centers, hist, alpha=1.)
ax.set_yscale('log')

img_gaussian_ifft = hpfn.get_phase_randomized(img_gaussian, rand_type='Gaussian')
hist, bin_centers = hpfn.get_hist(np.real(img_gaussian_ifft))

ax = fig.add_subplot(gs[0, 3])
ax.plot(bin_centers, hist, alpha=1.)
ax.set_yscale('log')

plt.show()
```

```python
contrast_collect = []
nat_numbers = np.load('/mnt/d/data/panoramic/natural_scenes_numbers.npy')

for ii in range(len(nat_numbers)):
    
    fig = plt.figure(figsize=(5, 10), constrained_layout=True)
    gs = GridSpec(5, 1, figure=fig)
    
    img = np.load(folder_nat + f'natural_{nat_numbers[ii]}.npy')
    K_row = img.shape[0] 
    K_col = img.shape[1] 

    shift = 0
    img_roll = np.roll(img, shift, axis=1)
    ax = fig.add_subplot(gs[0, 0])
    im = ax.imshow(img_roll,'gray')
    plt.colorbar(im, shrink=0.5)

    ####### phase randomized
    img_ifft = hpfn.get_phase_randomized(img)
    ax = fig.add_subplot(gs[1, 0])
    im = ax.imshow(np.real(img_ifft),'gray')
    plt.colorbar(im, shrink=0.5)

    # Spacial filtering
    pix_per_deg = K_col / 360
    FWHM = 5 # in degree
    sigma_for_gaussian = np.round(FWHM/(2*np.sqrt(2*np.log(2)))*pix_per_deg, 1) # smoothing gaussian
    pad_size = int(4*sigma_for_gaussian) # this comes from the fact that the gaussian is truncated at 4*std
    img_filtered = hpfn.get_filtered_spacial(np.real(img_ifft), pad_size, sigma_for_gaussian)
    ax = fig.add_subplot(gs[2, 0])
    im = ax.imshow(img_filtered,'gray')
    plt.colorbar(im, shrink=0.5)

    # resizing
    img_resized = hpfn.get_resized(img_filtered)
    ax = fig.add_subplot(gs[3, 0])
    im = ax.imshow(img_resized,'gray')
    plt.colorbar(im, shrink=0.5)

    # contrast
    scope = 'full'
    img_contrast = hpfn.get_contrast(img_resized.reshape((1, )+img_resized.shape), scope)
    ax = fig.add_subplot(gs[4, 0])
    im = ax.imshow(img_contrast[0],'gray')
    plt.colorbar(im, shrink=0.5)
    
    plt.savefig(f'../results/natural_phase_randomized_{nat_numbers[ii]}_{scope}.png', bbox_inches='tight')
    plt.close()
```
