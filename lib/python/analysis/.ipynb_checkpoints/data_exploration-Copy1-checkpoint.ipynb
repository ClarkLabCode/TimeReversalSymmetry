#### Look at the distributions of the randomized phase scenes

```python
contrast_collect = []
contrast_collect_normalized = []
nat_numbers = np.load('/mnt/d/data/panoramic/natural_scenes_numbers.npy')

for ii in range(len(nat_numbers)):
    img = np.load(folder_nat + f'natural_{nat_numbers[ii]}.npy')
    K_row = img.shape[0] 
    K_col = img.shape[1] 

    shift = 0
    img_roll = np.roll(img, shift, axis=1)

    ####### phase randomized
    img_ifft = hpfn.get_phase_randomized(img, rand_type='Gaussian' , delete_zero_freq=True)
    contrast_collect.extend(np.real(img_ifft.flatten()))

    # Spacial filtering
    pix_per_deg = K_col / 360
    FWHM = 5 # in degree
    sigma_for_gaussian = np.round(FWHM/(2*np.sqrt(2*np.log(2)))*pix_per_deg, 1) # smoothing gaussian
    pad_size = int(4*sigma_for_gaussian) # this comes from the fact that the gaussian is truncated at 4*std
    img_filtered = hpfn.get_filtered_spacial(np.real(img_ifft), pad_size, sigma_for_gaussian)

    # resizing
    img_resized = hpfn.get_resized(img_filtered)

    # contrast or normalization
#     scope = 'full'
#     img_contrast = hpfn.get_contrast(img_resized.reshape((1, )+img_resized.shape), scope)
    img_contrast = img_resized / abs(img_resized).max()
    
    contrast_collect_normalized.extend(img_contrast.flatten())
```

```python
fig = plt.figure(figsize=(12, 3), constrained_layout=True)
gs = GridSpec(1, 4, figure=fig)

hist, bin_centers = hpfn.get_hist(np.array(contrast_collect))

ax = fig.add_subplot(gs[0, 0])
ax.plot(bin_centers, hist, alpha=1.)
# plt.xlim([-100, 100])
ax.set_yscale('log')
ax.set_xlabel('luminance')
ax.set_ylabel('density')

hist, bin_centers = hpfn.get_hist(np.array(contrast_collect_normalized))
mean1 = np.round(np.array(contrast_collect_normalized).mean(), 2)
std1 = np.round(np.array(contrast_collect_normalized).std(), 2)
skew1 = np.round(skew(np.array(contrast_collect_normalized)), 2)
kurtosis1 = np.round(kurtosis(np.array(contrast_collect_normalized)), 2)

ax = fig.add_subplot(gs[0, 1])
ax.plot(bin_centers, hist, alpha=1.)
# plt.xlim([-100, 100])
ax.set_yscale('log')
ax.set_xlabel('contrast')
ax.set_ylabel('density')
ax.set_title(str(mean1) + ', ' + str(std1) + ', ' + str(skew1) + ', ' + str(kurtosis1), fontsize=10)

img_gaussian = np.random.normal(loc=0., scale=1, size=img.shape)
hist, bin_centers = hpfn.get_hist(np.real(img_gaussian))

ax = fig.add_subplot(gs[0, 2])
ax.plot(bin_centers, hist, alpha=1.)
ax.set_yscale('log')

img_gaussian_ifft = hpfn.get_phase_randomized(img_gaussian, rand_type='Gaussian')
hist, bin_centers = hpfn.get_hist(np.real(img_gaussian_ifft))

ax = fig.add_subplot(gs[0, 3])
ax.plot(bin_centers, hist, alpha=1.)
ax.set_yscale('log')

plt.show()

```

## Moving edges

```python
importlib.reload(hpfn)
# generate moving edge stimuli
K_row = 1
K_col = 360
pix_per_deg = K_col / 360
FWHM = 5 # in degree
sigma_for_gaussian = np.round(FWHM/(2*np.sqrt(2*np.log(2)))*pix_per_deg, 1) # smoothing gaussian
T = 50
delta_t = 0.01 # simulation step is 0.01 second
N_images = 1000
rescalev = 0.5

synthetic_data_white_edge = np.zeros((N_images, 5, 50, 1, 72))
synthetic_data_edge_targets = np.zeros((N_images, 1))

ind = 0
for ii in range(N_images):
    vel = np.ones(T-1) * (50 + np.random.random() * (200 - 50))
    min_dist = int(vel[0] * pix_per_deg * 0.5) + 1
    edge_pos = np.random.randint(K_col-min_dist)

    synthetic_data_edge_targets[ii, :] = vel[0]   

    # white edge
    img = -rescalev * np.ones((K_row, K_col))
    _, shift_array_pix = hpfn.get_shift_array(vel, delta_t, img)

    for tt, shift in enumerate(shift_array_pix):
        img[:, :edge_pos+shift] = rescalev
        img_filtered = hpfn.get_filtered_spacial_row(img, sigma_for_gaussian, 'reflect')
        for xx in range(5):
            synthetic_data_white_edge[ii, xx, tt, :, :] = img_filtered[:, xx::5]
synthetic_data_white_edge = synthetic_data_white_edge.reshape((N_images*5, 50, 1, 72))
synthetic_data_black_edge = -synthetic_data_white_edge
    
        
```

```python
fig = plt.figure(figsize=(12, 4), constrained_layout=True)
gs = GridSpec(2, 5, figure=fig)

for ii in range(5):
    ax = fig.add_subplot(gs[0, ii])
    im = ax.imshow(synthetic_data_white_edge[ii*10].squeeze(), 'gray')
    plt.colorbar(im)
    
    ax = fig.add_subplot(gs[1, ii])
    im = ax.imshow(synthetic_data_black_edge[ii*10].squeeze(), 'gray')
    plt.colorbar(im)

plt.show()
```

#### Check the statistics of the synthetic stimuli

```python
def get_statisics(train_samples):
    h1, bins1 = np.histogram(train_samples.flatten(), bins=int(np.sqrt(len(train_samples.flatten()))))
    mean1 = np.round(train_samples.flatten().mean(), 2)
    std1 = np.round(train_samples.flatten().std(), 2)
    skew1 = np.round(skew(train_samples.flatten()), 2)
    kurtosis1 = np.round(kurtosis(train_samples.flatten()), 2)

    return h1, bins1, mean1, std1, skew1, kurtosis1
```

```python
# gradient
importlib.reload(sawtooth_gradient)

K_col = 720
pix_per_deg = K_col / 360
FWHM = 5 # in degree
sigma_for_gaussian = np.round(FWHM/(2*np.sqrt(2*np.log(2)))*pix_per_deg, 1) # smoothing gaussian

scal = 0.5
stimr_all = []
for xtess in [8]:
    stimr = sawtooth_gradient.get_bitmap(direction='r', xtess=xtess, x=720, t=50)
    stimr = hpfn.get_filtered_spacial_row(stimr, sigma_for_gaussian)
    stimr = stimr[:, ::10] * scal
    stimr_all.append(stimr)
stimr_all = np.array(stimr_all)


fig = plt.figure(figsize=(4, 4), constrained_layout=True)
gs = GridSpec(1, 1, figure=fig)

h1, bins1, mean1, std1, skew1, kurtosis1 = get_statisics(stimr_all.flatten())
ax = fig.add_subplot(gs[0, 0])
ax.plot(bins1[1:], h1, c='k', alpha=0.5)
# ax.set_yscale('log')
ax.set_title(str(mean1) + ', ' + str(std1) + ', ' + str(skew1) + ', ' + str(kurtosis1), fontsize=20)
```
